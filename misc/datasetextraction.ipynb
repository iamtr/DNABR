```python
from Bio import SeqIO
import pandas as pd
```

```python
# Read the fasta file using Biopython
with open("TEST.fasta", "r") as file:
    records = list(SeqIO.parse(file, "fasta"))

data = []
for record in records:
    if len(record.seq) % 3 == 0:
        seq_length = len(record.seq)
        div = seq_length // 3
        seq = record.seq[:div - 1]
        dna_label = record.seq[div:div*2]
        rna_label = record.seq[div*2:]
        data.append({
            'id': record.id,
            'seq': str(seq),
            'dna_label': str(dna_label),
            'rna_label': str(rna_label)
        })
    
    # sanity check that protein sequence does not contain numbers
    if any(char.isdigit() for char in seq):
        print("The protein sequence contains numbers!")

df = pd.DataFrame(data)
print(df)
df.to_csv('DRNA_TEST.csv', index=False)
```
